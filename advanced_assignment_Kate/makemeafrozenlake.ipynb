```python
import numpy as np
```

```python
import gym

#create a single game instance
env = gym.make("FrozenLake-v0")

#start new game
env.reset();
```

```python
# display the game state
env.render()
```

```python
n_epochs = 1000

t_max = 25

agent = QLearningAgent(0.1,0.1,0.9,get_actions)
```

```python
def game() :
    reward = 0
    s = env.reset()
    
    agent.epsilon = agent.epsilon*0.99999
    
    for t in range(t_max):
        a = agent.getAction(s)
        pr_s = s
        s,r,done,_ = env.step(a)
        
        reward += r
        
        agent.update(pr_s,a,s,reward)
        
        if done : break
    
    return reward
```

```python
def evaluate(n_times=1000):
    rewards = [game() for i in range(n_times)]
    return float(np.mean(rewards))
```

```python

for epoch in range(n_epochs):
    print ("Epoch %s:"%epoch)
    print(agent.epsilon)
    
    print(evaluate())
```
